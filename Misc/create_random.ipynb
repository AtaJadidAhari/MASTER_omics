```r
import pandas as pd
```

```r
sa = pd.read_csv("/omics/odcf/analysis/hipo/hipo_021/outlier_analysis/sample_data/master_drop_sample_annotation_sizeFactorFiltered_0.1.tsv", sep="\t")

protein_sa = pd.read_csv("/omics/odcf/analysis/hipo/hipo_021/outlier_analysis/sample_data/protrider_sa.tsv", sep="\t")
```

```r
union_sa = list(set(sa["pid"]).union(set(protein_sa["pid"])))
len(union_sa)
```

```r
import random
import string

def generate_pseudo_id():
    # Generates 3 random uppercase letters
    letters = ''.join(random.choices(string.ascii_uppercase, k=3))
    # Generates 4 random digits
    numbers = ''.join(random.choices(string.digits, k=4))
    return f"{letters}{numbers}"

def anonymize_list(name_list):
    name_map = {}
    anonymized_data = []

    for name in name_list:
        if name not in name_map:
            # Create a unique ID if the name is new
            new_id = generate_pseudo_id()
            # Ensure the ID isn't a duplicate by chance
            while new_id in name_map.values():
                new_id = generate_pseudo_id()
            name_map[name] = new_id
        
        anonymized_data.append(name_map[name])
    
    return anonymized_data, name_map

# --- Usage ---
samples = ["Alice", "Bob", "Charlie", "Alice", "Diana"]
ids, mapping_key = anonymize_list(union_sa)

mapping_key
```
